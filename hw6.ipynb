Завдання 1

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

dataset_2d = pd.read_csv("data_2d.csv")
X_2d = dataset_2d.values

inertia_2d = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_2d)
    inertia_2d.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia_2d, marker='o')
plt.title('Ліктевий метод для двовимірного датасету')
plt.xlabel('Кількість кластерів')
plt.ylabel('Inertia')
plt.show()

optimal_k_2d = 3  
kmeans_2d = KMeans(n_clusters=optimal_k_2d, random_state=42)
labels_2d = kmeans_2d.fit_predict(X_2d)

mnist = datasets.fetch_openml('mnist_784')
X_mnist = mnist.data.astype('float64')
y_mnist = mnist.target.astype('int')

inertia_mnist = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_mnist)
    inertia_mnist.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia_mnist, marker='o')
plt.title('Ліктевий метод для mnist датасету')
plt.xlabel('Кількість кластерів')
plt.ylabel('Inertia')
plt.show()

optimal_k_mnist = 10  
kmeans_mnist = KMeans(n_clusters=optimal_k_mnist, random_state=42)
labels_mnist = kmeans_mnist.fit_predict(X_mnist)
```

Завдання 2

```python
plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels_2d, cmap='viridis', edgecolor='k')
plt.title('Результат кластеризації для двовимірного датасету')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.show()

pca = PCA(n_components=2)
X_mnist_pca = pca.fit_transform(X_mnist)

plt.scatter(X_mnist_pca[:, 0], X_mnist_pca[:, 1], c=labels_mnist, cmap='viridis', edgecolor='k')
plt.title('Результат кластеризації для mnist датасету (після PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
